### Ejercicio 1

Representa en un único gráfico las funciones arcoseno y arcocoseno en el intervalo [-pi/4, pi/4] 

```python
%matplotlib inline

# Respuesta
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

# Calculamos el array con un paso de 0.1
x = np.arange(-1*np.pi/4, 1*np.pi/4, 0.1)

# Array x, cos(x), 'r' hará utilizar el color rojo (red), array x, sin(x) y verde (green)
plt.plot(x, np.arccos(x), 'r', x, np.arcsin(x), 'g')

# Añadimos los nombres a los ejes x e y respectivamente:
plt.xlabel('x')
plt.ylabel('cos(x)')

# Mostramos el gráfico
plt.show()
```

### Ejercicio 2

Representa en un gráfico 3D la función `f(x,y) = cos(x) + 2*sen(y)` en el intervalo [-2*pi, 2*pi]

```python
%matplotlib inline

from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np
import math

fig = plt.figure()
ax = plt.axes(projection="3d")

def z_function(x, y):
    return np.cos(x)+2*np.sin(y)

x = np.linspace(-2*math.pi, 2*math.pi, 20)
y = np.linspace(-2*math.pi, 2*math.pi, 20)

X, Y = np.meshgrid(x, y)
Z = z_function(X, Y)

ax.plot_wireframe(X, Y, Z, color='green')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```

### Ejercicio 4

Dados los siguientes datos de meteorología registrados en el aeropuerto de Barcelona, representa en uno o varios gráficos los valores de temperatura media anual, máxima, mínima, media de lluvias mensual y humedad relativa.

Las posiciones del vector `meteo` son las siguientes:

* Nombre mes
* Temperatura media anual del mes (°C)
* Temperatura máxima media (°C)
* Temperatura mínima media (°C)
* Media de lluvias mensual (mm)
* Humedad media relativa (%)


```python
%matplotlib inline

meteo = [["January", 9.2, 13.6, 4.7, 37, 70],
["February",9.9, 14.3, 5.4, 35, 70],
["March", 11.8, 16.1, 7.4, 36, 70],
["April", 13.7, 18.0, 9.4, 40, 69],
["May", 16.9, 21.1, 12.8, 47, 70],
["June", 20.9, 24.9, 16.8, 30, 68],
["July", 23.9, 28.0, 19.8, 21, 67],
["August", 24.4, 28.5, 20.2, 62, 68],
["September", 21.7, 26.0, 17.4, 81, 70],
["October", 17.8, 22.1, 13.5, 91, 73],
["November", 13.0, 17.3, 8.6, 59, 71],
["December", 10.0, 14.3, 5.7, 40, 69]]

# Respuesta
month = []
average_annual_of_temperature_month = []
average_maximum_temperature = []
average_minimum_temperature = []
average_monthly_rainfall = []
relative_humidity = []

for all_ in meteo:
    month.append(all_[0])
    average_annual_of_temperature_month.append(all_[1])
    average_maximum_temperature.append(all_[2])
    average_minimum_temperature.append(all_[3])
    average_monthly_rainfall.append(all_[4])
    relative_humidity.append(all_[5])

    
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

#GRAFICO 1
#Medias en grados
plt.title("Average annual temperature of the month")
#Le doy la variable x a meses y la y a las medias de temperaturas (diferentes para que haga 3 curvas -> 3 medias)
x = month
y1 = average_annual_of_temperature_month
y2 = average_maximum_temperature
y3 = average_minimum_temperature
#la expreso en la grafica y decido el color
plt.plot(y1, 'pink')
plt.plot(y2, 'purple')
plt.plot(y3, 'orchid')
#controlo lo que quiero que me aparezca en el eje x de variable y su rotacion (para que quede girado)
plt.xticks(np.arange(len(month)), month, rotation=40)
plt.ylabel('temperatures(ºC)') 
plt.show()


#GRAFICO 2
plt.title("Average monthly of rainfall")
x = month
y = average_monthly_rainfall
plt.plot(y, 'r')
plt.xticks(np.arange(len(month)), month, rotation=40)
plt.ylabel('temperatures(mm)') 
plt.show()

#GRAFICO 3
plt.title("Relative humidity")
x = month
y = relative_humidity
plt.plot(y, 'yellowgreen')
plt.xticks(np.arange(len(month)), month, rotation=40)
plt.pie(relative_humidity, labels = month, startangle=90, shadow=True, autopct='%1.1f%%')  # Dibuja un gráfico de quesitos
plt.show()
```
